ðŸ”¹ STEP 4: Temporal Reduction â€“ Summarize 56 steps â†’ 1 row For each sequence_id, reduce to:

mean, std, min, max per column

```python
def extract_features(df, is_test=False):
    features = []
    for seq_id, group in df.groupby('sequence_id'):
        stats = group.describe().loc[['mean', 'std', 'min', 'max']]
        flat = stats.stack().to_frame().T
        flat.columns = [f"{feature}_{stat}" for stat, feature in flat.columns]
        flat['sequence_id'] = seq_id
        if not is_test:
            flat['gesture'] = group['gesture'].iloc[0]
        features.append(flat)
    return pd.concat(features, ignore_index=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from joblib import dump

# Step 1: Extract features from the cleaned DataFrame
features_df = extract_features(train_df)

# Step 2: Define feature matrix X and target vector y
X = features_df.drop(columns=['sequence_id', 'gesture'])
y = features_df['gesture']




# Remove rare gestures with <2 samples
valid_classes = y.value_counts()[y.value_counts() >= 2].index
X = X[y.isin(valid_classes)]
y = y[y.isin(valid_classes)]

# Split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Train
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Save
dump(clf, "model.joblib")
print("âœ… model.joblib saved!")
```
